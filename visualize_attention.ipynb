```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import math
import seaborn as sns
```

```python
def heat_map(att,color='red'):
    words = []
    y = []
    for a,b in att:
        words.append(a)
        y.append(b)
    y = np.expand_dims(np.array(y), axis=0)
    fig, ax = plt.subplots(figsize=(20,0.5))
    sns.set(font_scale=1.3)
    c = 'red'
    if color == 'red':
        c = 'OrRd'
    elif color == 'blue':
        c = 'Blues'
    elif color == 'green':
        c = 'Greens'
    elif color == 'purple':
        c = 'BuPu'
    sns.heatmap(y,xticklabels=words, cmap=c, linewidths=1.5)
```
